# Modulación en Amplitud

$$ e(t) = A \cdot [1 + m\cdot f(t)] \cos(w_p\cdot t)$$

$$ f(t) = moduladora $$

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
def modulada_amplitud(t,moduladora,m,wp,A):
    """ Dadas:
    t: tiempo (real)
    moduladora: función f(t).
    m: indice de modulacion (real).
    wp: frecuancia de portadora (real).
    A: amplitud de portadora (real).
    devuelve:
    el valor de la señal modulada en t(real)"""
    
    return A * (1 + m * moduladora(t)) * np.cos(wp * t)
```

```python
def moduladora_1(t):
    return np.sin(5 * t)

def moduladora_2(t):
    return np.sin(10 * t) * np.cos(2 * t)

def moduladora_3(t):
    return np.sin(15 * t)* np.cos(10 * t) ** 2

def moduladora_4(t):
    return np.sin(10 * t) * np. sin(7 * t)
```

```python
t = np.linspace(0,10,1000)

A = 1
m = 0.7
w_p = 20 * np.pi

moduladora = moduladora_1

onda_modulada = modulada_amplitud(t, moduladora, m, w_p, A)

fig, ax = plt.subplots(3, 1)

fig.set_figheight(8)
fig.set_figwidth(15)

ax[0].plot(t, moduladora(t))
ax[0].set_title('Moduladora')

ax[1].plot(t, A * np.cos(w_p * t))
ax[1].set_title('Portadora')

ax[2].plot(t, onda_modulada)
ax[2].set_title('Modulada')
```

# ¿no es muy "cutre" tener que cambiarlo a mano todo el rato?

```python
from IPython.html.widgets import interact
```

```python
def modulacion_widgets(m, w_p, A, n):
    
    N = (2 * np.pi * w_p) * 10
    
    t = np.linspace(0,10,int(N))
    
    moduladoras = [moduladora_1, moduladora_2, moduladora_3, moduladora_4]
    
    onda_modulada = modulada_amplitud(t, moduladoras[n-1], m, w_p, A)

    fig, ax = plt.subplots(3, 1)

    fig.set_figheight(6)
    fig.set_figwidth(15)

    ax[0].plot(t, moduladoras[n-1](t))
    ax[0].set_title('Moduladora %i' %n)

    ax[1].plot(t, A * np.cos(w_p * t))
    ax[1].set_title('Portadora')

    ax[2].plot(t, onda_modulada)
    ax[2].set_title('Modulada')
```

```python
interact(modulacion_widgets, m=(0.,1.), A=(0,10.), w_p=(5,100), n=(1,4))
```
